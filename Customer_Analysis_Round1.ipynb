### Correct the values in the column customer lifetime value. They are given as a percent, so multiply them by 100 and change dtype to numerical type:

```python
#Let's get some information about the columns of the DataFrame:

datasafe.info()
```

```python
datasafe.head(5)
```

As we can see the column Data Type is Object, therefore we will not be able to multiply the values straightaway. First of all, we will remove the percentage symbols out of those cells and after we will change the values to Numeric.

```python
def percentage_remover(x):
    '''
    This formula removes percetage symbols out of strings.
    '''

    if (isinstance(x,str) ):
        x = x.replace("%","")
        x = float(x)
    else: 
        x = float(x)
    return x
```

```python
#Creating an additional copy of the DataFrame to observe 
datasafe2 = datasafe.copy()

#Applying the function on the desired column of the DataFrame
datasafe2["customer lifetime value"] = datasafe2["customer lifetime value"].apply(percentage_remover)
datasafe2
```

```python
#Converting the values of the column from Object to Numerical:

datasafe2["customer lifetime value"] = pd.to_numeric(datasafe2["customer lifetime value"], errors='coerce')
display(datasafe2.head())
```

```python
datasafe2[["customer lifetime value"]]
```

```python
def multiply(x):
    '''this function multiplies the input times 100'''
    return x*100
```

```python
#Applying the function to the desired column of the DataFrame:
datasafe2["customer lifetime value"] = list(map(multiply, datasafe2["customer lifetime value"]))
display(datasafe2.head())
```

We can see that the transformation has been done successfully
