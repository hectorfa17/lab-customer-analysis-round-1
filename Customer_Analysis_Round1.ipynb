```python
#Correct the values in the column customer lifetime value. 
#They are given as a percent, so multiply them by 100 and change dtype to numerical type.
```

```python
def clean_clv(x):
    if (isinstance(x,str) ):
        x = x.replace("%","")
        x = float(x)
    else: 
        x = float(x)
    return x
```

```python
datasafe["Customer Lifetime Value"] = list(map(clean_clv, datasafe["Customer Lifetime Value"]))
display(datasafe.head())
```

```python
datasafe['Customer Lifetime value'] = pd.to_numeric(datasafe['Customer Lifetime Value'], errors='coerce')
display(datasafe.head())
datasafe.info()
```

```python
def multiply(x):
    return x*100
```

```python
datasafe["Customer Lifetime Value"] = list(map(multiply, datasafe["Customer Lifetime Value"]))
display(datasafe.head())
```
